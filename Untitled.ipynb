## 補值

```python
import pandas as pd
import numpy as np
df = pd.read_json('contacts.json')
df
```

```python
email = pd.Series(np.array(df['Email'].sort_values().unique())).to_list()
del email[0]
phone = pd.Series(np.array(df['Phone'].sort_values().unique())).to_list()
del phone[0]
orderid = pd.Series(np.array(df['OrderId'].sort_values().unique())).to_list()
del orderid[0]
# df['Id'] = str(df['Id'])
```

```python
df['same_email'] = df.groupby('Email')['Contacts'].transform('sum')
df['same_phone'] = df.groupby('Phone')['Contacts'].transform('sum')
df['same_orderid'] = df.groupby('OrderId')['Contacts'].transform('sum')
```
